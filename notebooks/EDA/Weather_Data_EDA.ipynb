```python
# load the standardized values not aggregated by state yet in case we want to do further analysis at location level
standardized_temperatures = pd.read_csv('temperatures_1981-01-01-2021-10-01.csv')#'standardized_temperatures.csv')
standardized_temperatures = standardized_temperatures.dropna()
standardized_droughts = pd.read_csv('pdsi_1981-01-01-2021-10-01.csv')
standardized_droughts = standardized_droughts.dropna()
# load the averages by state
#temperature_mean_grouped = pd.read_csv('standardized_temperatures_state.csv')
#drought_mean_grouped = pd.read_csv('standardized_droughts_state.csv')
# load the means and standard deviations by location
temp_summary = pd.read_csv('../merge_ready/annual_weather_final.csv')#'temperature_summary.csv')
temp_summary = temp_summary.dropna()
drought_summary = pd.read_csv('../merge_ready/annual_weather_final.csv')#'drought_summary.csv')
drought_summary = drought_summary.dropna()
```

```python
# convert the data into F for ease of users
for col in temp_summary:
  if 'Averages' in col:
    temp_summary[col+' F'] = (temp_summary[col]-273.15)*9/5+32
  elif 'Standard Deviation' in col:
    temp_summary[col+' F'] = temp_summary[col]*9/5
```

```python
# to show the hotter than average months (atleast averaged across the country)
print(standardized_temperatures.copy().mean().sort_values().tail(8))
# to show the colder than average months (atleast averaged across the country)
print(standardized_temperatures.copy().mean().sort_values().head(8))
```

```python
hot_summer = '2020-08-01 00:00:00'
cold_winter = '2010-02-01 00:00:00'
```

```python
fig = (px.scatter_mapbox(standardized_temperatures[[hot_summer,'x','y']], lat="y", lon="x", #hover_name=hover_name, 
                         #hover_data= text_data + generation_cols,
                         color=hot_summer,
                         zoom=2, height=600,opacity=0.5)
        )

fig.update_layout(mapbox_style="open-street-map")
#fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
